```python
def read_files_into_dict_lowercased(file_paths):
    file_contents = {}

    for file_path in file_paths:
        with open(file_path, 'r') as file:
            file_content = file.read().lower()  # Convert content to lowercase
            file_contents[file_path] = file_content

    return file_contents
```

**Reading All File Content Once:**

Efficiency: Opening and closing files repeatedly can be less efficient, especially if you're working with many files. Reading all the content at once and storing it in memory reduces the overhead of file I/O operations.

Simplicity: It simplifies the code by separating the file reading process from the data processing. Once you have all the content in memory, you can perform various operations without worrying about file access.

**Using .lower() Function:**

Standardization: Converting the text to lowercase is a common preprocessing step in text analysis and text mining. It ensures that text is treated uniformly and doesn't differentiate between uppercase and lowercase letters.

Consistency: When working with text data, making it all lowercase ensures that you don't have to deal with case sensitivity, making comparisons and analysis more consistent.


# **Count the number of characters in each text files.**

```python
def count_characters_in_contents(contents):
    character_counts = {}

    for file_path, file_content in contents.items():
        character_count = len(file_content)
        character_counts[file_path] = character_count

    return character_counts
```

```python
character_counts = count_characters_in_contents(english_file_contents)
for file_path, count in character_counts.items():
    print(f"File '{file_path}' contains {count} characters.")
```

# **Count the number of letters in each text files.**

```python
def count_letters_in_contents(contents):
    letter_counts = {}

    for file_path, file_content in contents.items():
        # Filter out non-alphabetic characters and count the letters
        letter_count = sum(1 for char in file_content if char.isalpha())
        letter_counts[file_path] = letter_count

    return letter_counts
```

```python
letter_counts = count_letters_in_contents(english_file_contents)
for file_path, count in letter_counts.items():
    print(f"File '{file_path}' contains {count} letters.")
```

# **Count the number of letters in each text files.**

```python
def count_letter_frequencies_in_contents(contents):
    letter_frequencies = {}

    for file_path, file_content in contents.items():
        letter_frequency = {}

        for char in file_content:
            if char.isalpha():
                char = char.lower()  # Convert to lowercase for case-insensitive counting
                if char in letter_frequency:
                    letter_frequency[char] += 1
                else:
                    letter_frequency[char] = 1

        letter_frequencies[file_path] = letter_frequency

    return letter_frequencies
```

```python
letter_frequencies = count_letter_frequencies_in_contents(english_file_contents)

for file_path, frequencies in letter_frequencies.items():
    print(f"Letter frequencies in file '{file_path}':")
    for letter, count in frequencies.items():
        print(f"'{letter}': {count} times")
```

# **Count the number of words in each text files.**

```python
def count_words_in_contents_using_split(contents):
    word_counts = {}

    for file_path, file_content in contents.items():
        words = file_content.split()  # Split content into words
        word_count = len(words)
        word_counts[file_path] = word_count

    return word_counts
```

```python
word_counts = count_words_in_contents_using_split(english_file_contents)

for file_path, count in word_counts.items():
    print(f"File '{file_path}' contains {count} words.")
```

```python
import re

def count_words_in_contents_using_regex(contents):
    word_counts = {}

    # Define a regex pattern for splitting on non-alphanumeric characters

    #[\W_]: This part of the pattern defines a character class that matches any non-alphanumeric character or an underscore.

    #+: The + symbol after the character class means "one or more occurrences."

    #It indicates that you want to split the string wherever there is one or more non-alphanumeric characters in a row.

    split_pattern = r'[\W_]+'

    for file_path, file_content in contents.items():
        words = re.split(split_pattern, file_content)
        words = [word for word in words if word]  # Remove empty strings
        word_count = len(words)
        word_counts[file_path] = word_count

    return word_counts
```

```python
word_counts_final = count_words_in_contents_using_regex(english_file_contents)

for file_path, count in word_counts.items():
    print(f"File '{file_path}' contains {count} words.")
```

# **Give the number of occurrences of each words detected in each file**

```python
import re
from collections import Counter

def count_word_occurrences_in_contents(contents):
    word_occurrences = {}

    # Define a regex pattern for splitting on non-alphanumeric characters
    split_pattern = r'[\W_]+'

    for file_path, file_content in contents.items():
        words = re.split(split_pattern, file_content)
        # Remove empty strings and convert to lowercase for case-insensitive counting
        words = [word.lower() for word in words if word]
        word_count = Counter(words)
        word_occurrences[file_path] = word_count

    return word_occurrences
```

```python
word_occurrences_in_each_file = count_word_occurrences_in_contents(english_file_contents)

# Display word occurrences in each file
for file_path, word_counts in word_occurrences_in_each_file.items():
    print(f"Word occurrences in '{file_path}':")
    for word, count in word_counts.items():
        print(f"'{word}': {count} times")
```

# **Give the number of occurrences of each words detected in all the files**

```python
import re
from collections import Counter

def count_word_occurrences_in_contents_for_each_file(contents):
    all_word_occurrences = Counter()

    # Define a regex pattern for splitting on non-alphanumeric characters
    split_pattern = r'[\W_]+'

    for file_path, file_content in contents.items():
        words = re.split(split_pattern, file_content)
        # Remove empty strings and convert to lowercase for case-insensitive counting
        words = [word.lower() for word in words if word]
        word_count = Counter(words)
        all_word_occurrences.update(word_count)

    return all_word_occurrences
```

```python
all_word_occurrences = count_word_occurrences_in_contents_for_each_file(english_file_contents)

print("Word occurrences across all files:")
for word, count in all_word_occurrences.items():
    print(f"'{word}': {count} times")
```

# **Build a list that represent the least significant words in the dataset (stop words).**

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')  # Download the stopwords dataset if you haven't already

def get_stop_words_in_vocabulary(vocabulary_df):
    # Get a list of NLTK's English stop words
    stop_words = set(stopwords.words('english'))

    # Find and return the stop words present in the vocabulary DataFrame
    stop_words_found = vocabulary_df[vocabulary_df['Word'].str.lower().isin(stop_words)]['Word'].tolist()

    return stop_words_found

def filter_stop_words_from_vocab(vocabulary_df):
    # Get a list of NLTK's English stop words
    stop_words = set(stopwords.words('english'))

    # Filter out stop words from the vocabulary DataFrame
    vocabulary_df_filtered = vocabulary_df[~vocabulary_df['Word'].str.lower().isin(stop_words)]

    return vocabulary_df_filtered
```

```python
import nltk
import pandas as pd
# Download the necessary NLTK data (if you haven't already)
nltk.download('punkt')

def count_words_in_files_nltk(file_contents):
    word_counts = {}

    for file_path, file_content in file_contents.items():
        # Tokenize the text into words using NLTK's word tokenizer
        words = nltk.word_tokenize(file_content)

        # Count the number of word tokens
        word_count = len(words)
        word_counts[file_path] = word_count

    return word_counts



def count_word_occurrences_in_files_nltk(file_contents):
    word_occurrences = {}

    for file_path, file_content in file_contents.items():
        # Tokenize the text into words using NLTK's word tokenizer
        words = nltk.word_tokenize(file_content)

        # Count word occurrences in the file using Counter
        word_counts = Counter(words)
        word_occurrences[file_path] = word_counts

    return word_occurrences


def count_word_occurrences_in_all_files_nltk(file_contents):
    all_word_occurrences = Counter()

    for file_content in file_contents.values():
        # Tokenize the text into words using NLTK's word tokenizer
        words = nltk.word_tokenize(file_content)

        # Update the word occurrences across all files using Counter
        all_word_occurrences.update(words)

    return all_word_occurrences


def build_vocabulary_table_nltk(word_occurrences):
    # Create a DataFrame to represent the vocabulary
    vocabulary_df = pd.DataFrame(list(word_occurrences.keys()), columns=['Word'])

    return vocabulary_df

def build_vocabulary_table_with_word_count_nltk(all_word_occurrences):
    # Create a DataFrame to represent the vocabulary
    vocabulary_df = pd.DataFrame(list(all_word_occurrences.items()), columns=['Word', 'Count'])

    return vocabulary_df


```
